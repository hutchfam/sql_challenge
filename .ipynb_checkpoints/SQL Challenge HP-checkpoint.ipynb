```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import sqlalchemy
from sqlalchemy import create_engine, Table
```

```python
engine = create_engine(f"postgresql://username:password@localhost:5432/sql_challenge")
conn = engine.connect()
```

```python
salaries=pd.read_sql("SELECT*FROM salaries", conn)
salaries.head()
```

```python
titles=pd.read_sql("SELECT*FROM titles", conn)
titles.columns = ['emp_title', 'title']
titles.head()
```

```python
employ=pd.read_sql("SELECT*FROM employees", conn)
employ.head()
```

```python
#common_salary = salaries['salary'].tolist()
#plt.hist(common_salary)
plt.title("common salary histogram")
plt.xlabel("Salary")
plt.ylabel("Employees")
plt.show()
```

```python
salary_employee_comb = pd.merge(employ, salaries, on="emp_no", how="outer")
salary_employee_comb.head()
```

```python
salary_employee_title_comb = pd.merge(salary_employee_comb, titles, on="emp_title", how="outer")
salary_employee_title_comb.head()
```

```python
x_axis=salary_employee_title_comb["title"]
y_axis= salary_employee_title_comb["salary"]
ticks = np.arange(len(x_axis))
plt.bar(x_axis, y_axis, color='r', alpha=0.5, align="center")
plt.label("Salaries by Title")
plt.ylabel("Salaries")
plt.xlabel("Title")
plt.show()
```
